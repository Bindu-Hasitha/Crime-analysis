```python
murderg = data.groupby(['Year' , 'Group_Name'])['Victims_Total'].sum().reset_index() # grouping with year and sub group
murderg = murderg[murderg['Group_Name']!= '3. Total']   # we dont need total category of sub group

plt.style.use("fivethirtyeight")
plt.figure(figsize = (8,7))
ax = sns.barplot( x = 'Year', y = 'Victims_Total' , hue = 'Group_Name' , data = murderg ,palette= 'bright') #plotting barplot
plt.title('Gender Distribution of Victims per Year',size = 10)
ax.set_ylabel('')
```
